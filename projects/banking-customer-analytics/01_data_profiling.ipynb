```python
columns_to_review = [
    "gender",
    "occupation",
    "origin_province",
    "married",
    "active_member",
    "exit",
    "customer_segment",
    "loyalty_level",
    "digital_behavior",
    "risk_segment",
    "cluster_group"
]

for column in columns_to_review:
    print(f"\n--- {column} ---")
    print(df[column].value_counts(dropna=False))
```

```python
text_columns = df.select_dtypes(include="object").columns

for column in text_columns:
    whitespace_count = (
        df[column]
        .dropna()
        .astype(str)
        .ne(df[column].dropna().astype(str).str.strip())
        .sum()
    )

    if whitespace_count > 0:
        print(f"{column}: {whitespace_count} values contain leading/trailing spaces")
```

## 9. Target Variable Distribution

```python
df["exit"].value_counts(dropna=False)
```

```python
target_summary = pd.DataFrame({
    "count": df["exit"].value_counts(dropna=False),
    "percentage": df["exit"].value_counts(
        normalize=True,
        dropna=False
    ) * 100
})

target_summary
```

```python
boolean_columns = df.select_dtypes(include="bool").columns

for column in boolean_columns:
    print(f"\n--- {column} ---")
    print(df[column].value_counts(dropna=False))
```

### Findings

The target variable (`exit`) is a Boolean field representing customer churn.

- **False:** Customer retained (65,600 records, 82.0%)
- **True:** Customer churned (14,400 records, 18.0%)

The dataset is moderately imbalanced, with churned customers representing 18% of the total population. This distribution should be considered in subsequent exploratory analysis and predictive modelling.

## 10. Date Variable Review

```python
for column in ["last_active_date", "created_date"]:
    parsed_dates = pd.to_datetime(df[column], errors="coerce")

    print(f"\n{column}")
    print(f"Invalid dates: {parsed_dates.isna().sum():,}")
    print(f"Earliest date: {parsed_dates.min()}")
    print(f"Latest date: {parsed_dates.max()}")
```

## 11. Basic Validity Checks

```python
validity_checks = {
    "age_below_zero": (df["age"] < 0).sum(),
    "age_above_100": (df["age"] > 100).sum(),
    "negative_balance": (df["balance"] < 0).sum(),
    "negative_monthly_income": (df["monthly_ir"] < 0).sum(),
    "negative_tenure": (df["tenure_ye"] < 0).sum(),
    "negative_card_count": (df["nums_card"] < 0).sum(),
    "negative_service_count": (df["nums_service"] < 0).sum()
}

pd.Series(validity_checks, name="record_count")
```

```python
(df["tenure_ye"] > df["age"]).sum()
```

```python
created_date = pd.to_datetime(df["created_date"], errors="coerce")
last_active_date = pd.to_datetime(df["last_active_date"], errors="coerce")

(created_date > last_active_date).sum()
```
